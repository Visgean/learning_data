```python
%matplotlib notebook

from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plot
import pandas as pd
```

```python
faith = pd.read_csv('faithful.txt', delimiter='\t')

X = faith.as_matrix()
X[:3]
```

```python
k_means = KMeans(n_clusters=2).fit_predict(X)
```

```python
plot.scatter(X[:, 0], X[:, 1], c=k_means)
```
